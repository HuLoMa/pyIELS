# Code calculating the classical Wigner function after propagation

### Imports

```python
%matplotlib widget

from src.wigner_function import wigner_classical
from src.wigner_methods import wigner_spectral_convolve, wigner_propagate
from src.plot_function import plot_wigner_electron
```

### Simulations inputs

```python
ene_photon=1.55 # Photon energy in eV
g=2.0 # Electron-light coupling constant
prop_mm=5.0 # propagation distance after interaction in mm
ZLP_width=0.7 # Energy standard deviation in eV
```

### Compute the Wigner function

```python
wigner_exact, sidebands_axis, z_axis = wigner_classical(ene_photon,g)
```

### Convolve the Wigner Function with a realistic ZLP

```python
wigner_convolved, p_axis = wigner_spectral_convolve(wigner_exact, ene_photon, ZLP_width)
```

### Propagate the Wigner Function after interaction

```python
wigner_propagated=wigner_propagate(wigner_convolved, p_axis, ene_photon, prop_mm)
```

### Plot the Wigner function together with its marginal distributions

```python
plot_wigner_electron(wigner_propagated, p_axis, z_axis, ene_photon)
```
